##Análise dos dados de Rondônia:

```
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Dados obtidos em https://brasil.io/dataset/covid19/caso/?state=RO&place_type=state&page=233
# UF = RO
# TipoDeLocal=state

dados = pd.read_csv("https://github.com/ctulio/agendamento-hospitalar/raw/main/covid19-5c565ef7b6ac4f109f939c622553cd05.csv")
print(dados.columns)
print(dados.shape)

# Organizando os dados e o índice para que os dados constem do mais antigo para o mais recente
dados = dados.sort_values("date")
dados = dados.reset_index(drop=True)

# Formatando a coluna data (YYYY-mm-dd -> dd/mm/YYYY)
dados["date"] = pd.to_datetime(dados.date)
dados["date"] = dados["date"].dt.strftime("%d/%m/%Y")

dados.head()
```

Foi observado que o dataset possui apenas o número acumulado de casos confirmados **(confirmed)**, ele não possui a informação de casos confirmados por dia.
Por isso, para obter a quantidade de casos confirmados por dia subtraí o número de casos de um dia pelo número de casos do dia anterior, essa nova coluna eu chamei de **confirmed_on_day**.

```
# Criando uma lista com a quantidade de casos confirmados por dia
confirmed_on_day = []
confirmados = dados["confirmed"].tolist()


for i in range(len(confirmados)):
  diferenca = confirmados[i] - confirmados[i-1]
  confirmed_on_day.append(diferenca)
print(f"Quantidade de itens: {len(confirmed_on_day)}")
print(confirmed_on_day[0:10])

# A operação acima gera o primeiro elemento a partir da subtração do primeiro pelo último ([0] - [-1])
# Por isso é necessário ajustar o primeiro elemento
confirmed_on_day.pop(0) # retirar o primeiro
confirmed_on_day.insert(0,1) # inserir o número "1" na primeira posição ([0]), afinal, no primeiro dia houve 1 caso confirmado

print(f" {confirmed_on_day[0:10]}")
```

```
# Criando a coluna de casos confirmados no dia (confirmed_on_day) no dataset
dados["confirmed_on_day"] = confirmed_on_day
print(dados.columns)
dados.head()
```

```
# Separando as colunas de interesse
evolucao = dados[["date", "confirmed_on_day"]]

# Fatiando o dataset para o período de campanha
periodo_campanha = evolucao.loc[190:]
print(periodo_campanha.head(2))
print(periodo_campanha.tail(2))

```

```
# Visualizando o período de campanha através do gráfico de barras
ax = periodo_campanha.plot(x="date", y="confirmed_on_day",kind="bar", figsize=(14,8), grid=True, label="Novos casos confirmados")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.xticks(rotation=45, horizontalalignment="right")
plt.legend().set_title('Legenda:')
plt.title("Variação de casos durante o perído de campanha")
plt.xlabel("Data (dia/mês/ano)", fontsize=15)
plt.ylabel("Número diário de casos confirmados", fontsize=15);
```

```
# Fatiando o dataset para o período anterior a campanha eleitoral
periodo_nao_campanha = evolucao.loc[142:190]
```

```
# Visualização do período anterior à campanha eleitoral
ax = periodo_nao_campanha.plot(x="date", y="confirmed_on_day", kind="bar", figsize=(14,8), grid=True, label="Novos casos confirmados")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.xticks(rotation=45, horizontalalignment="right")
plt.title("Variação de casos anterior ao período de campanha ");
```


A visualização através do gráfico de barras não trouxe indicações significativas.

Outra possibilidade de busca de insights seria visualizar as informações estatísticas que o método describe() nos traz.


```
periodo_campanha.describe()
```

```
periodo_nao_campanha.describe()
```

Em números absolutos o período anterior a campanha possui números maiores de casos, o que poderia induzir a uma conclusão de que o atual momento está melhor.

No entanto, não nos aponta sobre tendências. Para buscar alguma indicação de tendência, utilizo os gráficos abaixo:

```
plt.figure(figsize=(20,16))

plt.subplot(2,1,1,)
plt.title('Variação do número de casos ANTES do período de campanha', fontsize=20).set_color('b')
plt.plot(periodo_nao_campanha["date"], periodo_nao_campanha["confirmed_on_day"], color="b")
plt.grid(True)
plt.xticks(rotation=45, horizontalalignment="right")
plt.ylabel('Número de infectados por dia', fontsize=18)

plt.subplot(2,1,2)
plt.title('Variação do número de casos DURANTE o período de campanha', fontsize=20).set_color('r')
plt.plot(periodo_campanha["date"], periodo_campanha["confirmed_on_day"], color="r")
plt.grid(True)
plt.xticks(rotation=45, horizontalalignment="right")
plt.ylabel('Número de infectados por dia', fontsize=18);
```

---
Tentando reduzir um viés de escala, segue uma visualização concatenada dos períodos. Onde é possível verificar que a tendência de queda não se manteve no período eleitoral, inclusive iniciou um tendência de alta. 

```
plt.figure(figsize=(30,16))
plt.title('Visualização concatenada', fontsize=20).set_color('black')
plt.plot(periodo_nao_campanha["date"], periodo_nao_campanha["confirmed_on_day"], color="b")
plt.plot(periodo_campanha["date"], periodo_campanha["confirmed_on_day"], color="r")
plt.grid(True)
plt.xticks(rotation=45, horizontalalignment="right")
plt.ylabel('Número de infectados por dia', fontsize=18);
```

## Análise dos dados do Brasil:

A seguir, tentei reproduzir a metodologia com um grupo populacional maior, no caso, dados de todos os estados do Brasil.

```
dados_br = pd.read_csv("https://github.com/ctulio/agendamento-hospitalar/raw/main/covid19-372aaa9b6ae14bcfaae610dcda0936f5.csv")

#Organizando os dados e o índice para que os dados constem do mais antigo para o mais recente
dados_br = dados_br.sort_values("date")
dados_br = dados_br.reset_index(drop=True)


dados_br["state"].unique()
```

```
# Removendo colunas desnecessárias 
dados_br.drop(columns=["is_last", "city_ibge_code","city","estimated_population", "death_rate","place_type","confirmed_per_100k_inhabitants"], inplace=True)
dados_br
```

```

# Somando o total de casos/mortes em função do dia
evolucao_br = dados_br.groupby(["date"]).sum()
print( evolucao_br.head(5))
print(evolucao_br.tail(5))
```

```
# Criando a lista[] com o número de casos confirmados por dia
confirmed_on_day_br = []
confirmados_br = evolucao_br["confirmed"].tolist()

for i in range(len(confirmados_br)):
  diferenca = confirmados_br[i] - confirmados_br[i-1]
  confirmed_on_day_br.append(diferenca)

# A operação acima gera o primeiro elemento da lista a partir da subtração do primeiro pelo último ([0] - [-1])
# Por isso é necessário ajustar o primeiro elemento
confirmed_on_day_br.pop(0) # retirar
confirmed_on_day_br.insert(0,1) # inserir o número "1" na primeira posição ([0]), afinal, no primeiro dia houve 1 caso confirmado
print(f"Tamanho: {len(confirmed_on_day_br)}; Itens da lista: {confirmed_on_day_br[0:10]}")
```

```
# Criando a coluna de casos confirmados "confirmed_on_day" no dataset
evolucao_br["confirmed_on_day_br"] = confirmed_on_day_br
```

Observando os últimos elementos da nova coluna (confirmed_on_day_br), percebi alguns valores negativos. Inicialmente pensei que eu havia errado na lógica do cálculo e revisei algumas vezes o algoritmo.
Olhando com mais cuidado percebi na coluna (confirmed) que deveria ser o total de casos confirmados no brasil, acontecia casos em que o número anterior era maior que o próximo. Percebi que essa coluna não se tratava de uma coluna com valores cumulativos crescentes.

```
evolucao_br.tail(15)
```

Após algumas rodadas de testes e revisões me ocorreu a possibilidade de que a soma de cada dia poderia não conter as 27 unidades da federação, por isso, aconteceria de alguns dias estarem com um valor abaixo do esperado.

```
# Foi efetuado o agrupamento por data e a contagem de quantos elementos (estados) estavam presentes
T = dados_br.groupby(["date"])["state"].count()
print(T.tail(30))
```

Como é possível observar, ocorreu a situação de que alguns dias não estão com 27 elementos, o que nos induz a crer que alguns estados não enviaram suas informações para a totalização. Afetando dessa forma o cômputo dos valores de casos confirmados e consequentemente de casos confirmados por dia.

Em virtude da situação acima mencionada, no gráfico abaixo é possivel observar como fica distorcido a variação de novos casos por dia no perído de campanha eleitoral

```
evolucao_br.reset_index(inplace=True)
periodo_campanha_br = evolucao_br.loc[222:]
ax = periodo_campanha_br.plot(x="date", y="confirmed_on_day_br", figsize=(14,8), grid=True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.xticks(rotation=45, horizontalalignment="right")
plt.title("Variação de casos no Brasil durante o perído de campanha");
```
